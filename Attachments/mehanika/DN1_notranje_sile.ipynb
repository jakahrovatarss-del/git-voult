# Domača naloga 1 — Notranje sile in navori v rezkalnem orodju

**Konzola** dolžine $L = 34{,}75\ \text{mm}$, obremenjena s tremi silami na polmeru $r = 17{,}5\ \text{mm}$:

| Sila | Smer | Vrednost |
|------|------|----------|
| $F_a$ | aksialna (x) | $-35{,}5\ \text{N}$ |
| $F_r$ | radialna (y) | $37{,}25\ \text{N}$ |
| $F_c$ | obodna (z) | $113\ \text{N}$ |

Metoda preseka → 6 notranjih količin: $N$, $T_n$, $T_b$, $M_t$, $M_b$, $M_n$

## 1. Uvoz knjižnic in podatki

```python
import numpy as np
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# ── Podatki ──────────────────────────────────────────────────────────
L  = 34.75   # mm  – dolžina stebla konzole
r  = 17.5    # mm  – polmer orodja
Fa = -35.5   # N   – aksialna sila (negativna = tlak)
Fr =  37.25  # N   – radialna sila
Fc = 113.0   # N   – obodna (rezalna) sila

print(f"Podatki: L={L} mm, r={r} mm, Fa={Fa} N, Fr={Fr} N, Fc={Fc} N")
```

## 1b. Skice problema in metode preseka

Zahtevane skice:
1. **Skica konzole** z vsemi silami in polmerom $r$
2. **Prosti diagram telesa** — desni del pri preseku $s$ z vsemi 6 notranjimi količinami ($N$, $T_n$, $T_b$, $M_t$, $M_b$, $M_n$)

## 2. Metoda preseka — izračun notranjih količin

Za presek na koordinati $s$ ($0 \le s \le L$) obravnavamo **desni del** (od preseka do prostega konca) in postavimo ravnovesje:

### Sile
$$N(s) = -F_a = 35{,}5\ \text{N} \quad \text{(konst.)}$$
$$T_n(s) = -F_r = -37{,}25\ \text{N} \quad \text{(konst.)}$$
$$T_b(s) = -F_c = -113\ \text{N} \quad \text{(konst.)}$$

### Navori
$$M_t(s) = -F_c \cdot r \quad \text{(torzija, konst.)}$$
$$M_b(s) = -F_r \cdot (L - s) \quad \text{(upogib ok. z, linearna)}$$
$$M_n(s) = -F_c \cdot (L - s) - F_r \cdot r \quad \text{(upogib ok. y, linearna)}$$

```python
s = np.linspace(0, L, 500)

# Notranje sile (konstantne vzdolž osi)
N  = -Fa * np.ones_like(s)
Tn = -Fr * np.ones_like(s)
Tb = -Fc * np.ones_like(s)

# Navori
Mt = -Fc * r * np.ones_like(s)       # torzija (konstantna)
Mb = -Fr * (L - s)                    # upogibni moment okoli z (linearen)
Mn = -Fc * (L - s) - Fr * r          # upogibni moment okoli y (linearen)

# Izpis mejnih vrednosti
print("=== Vrednosti pri vpetju (s=0) ===")
print(f"  N   = {N[0]:+.2f} N")
print(f"  Tn  = {Tn[0]:+.2f} N")
print(f"  Tb  = {Tb[0]:+.2f} N")
print(f"  Mt  = {Mt[0]:+.2f} N·mm")
print(f"  Mb  = {Mb[0]:+.2f} N·mm")
print(f"  Mn  = {Mn[0]:+.2f} N·mm")
print()
print("=== Vrednosti na prostem koncu (s=L) ===")
print(f"  Mt  = {Mt[-1]:+.2f} N·mm")
print(f"  Mb  = {Mb[-1]:+.2f} N·mm")
print(f"  Mn  = {Mn[-1]:+.2f} N·mm")
```

## 3. Diagrami notranjih sil in navorov

```python
fig = plt.figure(figsize=(14, 16))
fig.suptitle(
    "Notranje sile in navori – rezkar (konzola)\n"
    f"$L = {L}$ mm,  $r = {r}$ mm,  "
    f"$F_a = {Fa}$ N,  $F_r = {Fr}$ N,  $F_c = {Fc}$ N",
    fontsize=13, fontweight="bold"
)

gs = gridspec.GridSpec(3, 2, figure=fig, hspace=0.55, wspace=0.35)

datasets = [
    (gs[0, 0], s, N,  "N(s)  [N]",        "Aksialna sila N",     "tab:blue"),
    (gs[0, 1], s, Tn, "$T_n$(s)  [N]",    "Strižna sila $T_n$",  "tab:orange"),
    (gs[1, 0], s, Tb, "$T_b$(s)  [N]",    "Strižna sila $T_b$",  "tab:green"),
    (gs[1, 1], s, Mt, "$M_t$(s)  [N·mm]", "Torzija $M_t$",       "tab:red"),
    (gs[2, 0], s, Mb, "$M_b$(s)  [N·mm]", "Upogibni moment $M_b$","tab:purple"),
    (gs[2, 1], s, Mn, "$M_n$(s)  [N·mm]", "Upogibni moment $M_n$","tab:brown"),
]

for spec, xs, ys, ylabel, title, color in datasets:
    ax = fig.add_subplot(spec)
    ax.plot(xs, ys, color=color, linewidth=2)
    ax.fill_between(xs, ys, 0, alpha=0.15, color=color)
    ax.axhline(0, color="black", linewidth=0.8)
    ax.set_xlabel("s  [mm]", fontsize=9)
    ax.set_ylabel(ylabel, fontsize=9)
    ax.set_title(title, fontsize=10, fontweight="bold")
    ax.set_xlim(0, L)

    y0, yL = float(ys[0]), float(ys[-1])
    yrange = float(np.ptp(ys)) if float(np.ptp(ys)) != 0 else max(abs(y0), abs(yL), 1)
    offset = 0.12 * yrange
    ax.text(0.5,     y0 + offset * (1 if y0 >= 0 else -1), f"{y0:.1f}",
            fontsize=8, ha="left",  color=color)
    ax.text(L - 0.5, yL + offset * (1 if yL >= 0 else -1), f"{yL:.1f}",
            fontsize=8, ha="right", color=color)
    ax.grid(True, alpha=0.3, linestyle="--")

plt.savefig("DN1_diagrami.png", dpi=150, bbox_inches="tight")
plt.show()
print("Slika shranjena: DN1_diagrami.png")
```

## 4. Verifikacija in skupni upogibni moment

```python
# Skupni upogibni moment vzdolž osi
M_up = np.sqrt(Mb**2 + Mn**2)

# Skupni upogibni moment – diagram
fig2, ax = plt.subplots(figsize=(8, 4))
ax.plot(s, M_up, color="navy", linewidth=2, label=r"$M_{up} = \sqrt{M_b^2 + M_n^2}$")
ax.fill_between(s, M_up, 0, alpha=0.15, color="navy")
ax.axhline(0, color="black", linewidth=0.8)
ax.set_xlabel("s  [mm]"); ax.set_ylabel("$M_{up}$  [N·mm]")
ax.set_title("Skupni upogibni moment $M_{up}(s)$", fontsize=11, fontweight="bold")
ax.set_xlim(0, L)
ax.text(0.5, M_up[0] * 1.03, f"{M_up[0]:.1f}", fontsize=9, color="navy")
ax.text(L - 0.5, M_up[-1] * 1.03, f"{M_up[-1]:.1f}", fontsize=9, ha="right", color="navy")
ax.grid(True, alpha=0.3, linestyle="--")
ax.legend()
plt.tight_layout()
plt.savefig("DN1_Mup.png", dpi=150, bbox_inches="tight")
plt.show()

# Numerični povzetek
print("=== Povzetek — kritično mesto (vpetje, s=0) ===")
print(f"  |Mb(0)|  = {abs(Mb[0]):.2f} N·mm")
print(f"  |Mn(0)|  = {abs(Mn[0]):.2f} N·mm")
print(f"  M_up(0)  = {M_up[0]:.2f} N·mm")
print(f"  |Mt|     = {abs(Mt[0]):.2f} N·mm  (torzija, konst.)")
print(f"\n  Razmerje torzija/upogib: Mt/M_up = {abs(Mt[0])/M_up[0]:.3f}")
```
